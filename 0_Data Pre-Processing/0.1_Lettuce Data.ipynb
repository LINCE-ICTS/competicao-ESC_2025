## "PEST_AI Plant Leaf Disease Recognition Dataset"
[Kaggle â€¢ _Ibrahima Gabar Diop_ Dataset](https://www.kaggle.com/datasets/ibrahimagabardiop/niayes-crops-disease-v2)

```python
# Replace with proper dataset reference
dataset_2 = "ibrahimagabardiop/niayes-crops-disease-v2"

custom_dir_2 = "../Datasets/lettuce_2"
# Download to a custom folder
api.dataset_download_files(dataset_2, path=custom_dir_2, unzip=True)
```

```python
import os
# Ensureing the directory structure is as expected

desired_plants = ["Lettuce__Bacterial", "Lettuce__fungal", "Lettuce__healthy"]
partitions = ["train", "test", "val"]

## List all the subfolders in the "val directory" of Train, Test and Validation data
train_subfolders = [
    # List subfolders inside Train partition
    name for name in os.listdir(os.path.join(custom_dir_2, partitions[0]))
    if os.path.isdir(os.path.join(custom_dir_2, partitions[0], name)) and name not in desired_plants
]
test_subfolders = [
    # List subfolders inside Test partition
    name for name in os.listdir(os.path.join(custom_dir_2, partitions[1]))
    if os.path.isdir(os.path.join(custom_dir_2, partitions[1], name)) and name not in desired_plants
]
val_subfolders = [
    # List subfolders inside Val partition
    name for name in os.listdir(os.path.join(custom_dir_2, partitions[2]))
    if os.path.isdir(os.path.join(custom_dir_2, partitions[2], name)) and name not in desired_plants
]
unwanted_subfolders = {"train": train_subfolders, "test": test_subfolders, "val": val_subfolders}
## Print the identified subfolders
print("Selected subfolders Length (train):", len(unwanted_subfolders["train"]))
print("Selected subfolders Length (test):", len(unwanted_subfolders["test"]))
print("Selected subfolders Length (val):", len(unwanted_subfolders["val"]))


import shutil
# Remove unwanted subfolders
for partition in partitions:
    for subfolder in unwanted_subfolders[partition]:
        subfolder_path = os.path.join(custom_dir_2, partition, subfolder)
        if os.path.exists(subfolder_path):
            shutil.rmtree(subfolder_path)
            print(f"Removed unwanted subfolder: {subfolder_path}")
```

```python
# Checking folders number of images in all test, train and valid folders
for i in partitions:
    print(f"\n--- {i} folder ---")
    base_path = os.path.join(custom_dir_2, i)

    for category in ["Lettuce__Bacterial", "Lettuce__fungal", "Lettuce__healthy", "diseased"]:
        category_path = os.path.join(base_path, category)
        if os.path.exists(category_path):
            num_files = len([f for f in os.listdir(category_path) if os.path.isfile(os.path.join(category_path, f))])
            print(f"{category}: {num_files} images")
        else:
            print(f"{category}: Folder does not exist")
```

# Data pre-processing

## "Lettuce plant Disease Dataset"
[Kaggle â€¢ _santosh shaha_ Dataset](https://www.kaggle.com/datasets/santoshshaha/lettuce-plant-disease-dataset)

### Adjusting folders

```python
# List for iterarion
for i in ["train", "test", "valid"]:
    print(f"\n--- Processing {i} folder ---")

    # Define the paths for the Bacterial and Fungal folders
    bacterial_folder = os.path.join(custom_dir_1, "lettuce", i, "Bacterial")
    fungal_folder = os.path.join(custom_dir_1, "lettuce", i, "fungal")

    # Create the diseased folder if it doesn't exist
    diseased_folder = os.path.join(custom_dir_1, "lettuce", i, "diseased")
    os.makedirs(diseased_folder, exist_ok=True)

    # Move all files from Bacterial and Fungal folders to diseased folder
    for folder in [bacterial_folder, fungal_folder]:
        if os.path.exists(folder):
            # Renaming Bacterial files and moving
            if folder == bacterial_folder:
                for filename in os.listdir(folder):
                    src = os.path.join(folder, filename)
                    dst = os.path.join(diseased_folder, f"bacterial_{filename}")
                    if os.path.isfile(src):
                        os.rename(src, dst)
            
            # Renaming Fungal files and moving
            elif folder == fungal_folder:
                for filename in os.listdir(folder):
                    src = os.path.join(folder, filename)
                    dst = os.path.join(diseased_folder, f"fungal_{filename}")
                    if os.path.isfile(src):
                        os.rename(src, dst)
```

```python
# Checking folders number of images in all test, train and valid folders
for i in ["train", "test", "valid"]:
    print(f"\n--- {i} folder ---")
    base_path = os.path.join(custom_dir_1, "lettuce", i)
    for category in ["Bacterial", "fungal", "healthy", "diseased"]:
        category_path = os.path.join(base_path, category)
        if os.path.exists(category_path):
            num_files = len([f for f in os.listdir(category_path) if os.path.isfile(os.path.join(category_path, f))])
            print(f"{category}: {num_files} images")
        else:
            print(f"{category}: Folder does not exist")
```

```python
# Dellete empty Bacterial and Fungal folders
import os
for i in ["train", "test", "valid"]:
    bacterial_folder = os.path.join(custom_dir_1, "lettuce", i, "Bacterial")
    fungal_folder = os.path.join(custom_dir_1, "lettuce", i, "fungal")

    for folder in [bacterial_folder, fungal_folder]:
        if os.path.exists(folder) and not os.listdir(folder):  # Check if folder exists and is empty
            os.rmdir(folder)
            print(f"Deleted empty folder: {folder}")
        else:
            print(f"Folder not deleted (either does not exist or is not empty): {folder}")
```
